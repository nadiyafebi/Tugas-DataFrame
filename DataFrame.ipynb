```python
import pandas as pd

# Load Data dari CSV
file_path = 'data_sampah.csv' 
df = pd.read_csv(file_path)

# Cetak nama kolom untuk memastikan sesuai
print("Nama kolom dalam dataset:", df.columns)
```

```python
# SOAL NO.1
print("\nJawaban No.1")
print("Data Awal:")
df.head()
```

```python
# SOAL NO.2: Total Sampah di Tahun 2016
tahun_target = 2016
total_sampah = df[df['tahun'] == tahun_target]['jumlah_produksi_sampah'].sum()

print("\nJawaban No.2 :")
print(f"Total sampah di seluruh Kabupaten/Kota pada tahun {tahun_target} adalah: {total_sampah:.2f} ton")

# Simpan hasil No.2 ke file CSV dan Excel
df_no2 = pd.DataFrame({'Tahun': [tahun_target], 'Total Sampah (ton)': [total_sampah]})
df_no2.to_csv('total sampah 2016.csv', index=False)
df_no2.to_excel('total sampah 2016.xlsx', index=False)
```

```python
# SOAL NO.3: Jumlah Sampah per Tahun
jumlah_sampah_per_tahun = df.groupby('tahun')['jumlah_produksi_sampah'].sum()

print("\nJawaban No.3 :")
print("Jumlah sampah per tahun:")
for tahun, total in jumlah_sampah_per_tahun.items():
    print(f"Tahun {tahun}: {total:.2f} ton")

# Simpan hasil No.3 ke file CSV dan Excel
df_no3 = jumlah_sampah_per_tahun.reset_index()
df_no3.columns = ['Tahun', 'Total Sampah (ton)']
df_no3.to_csv('jumlah sampah pertahun.csv', index=False)
df_no3.to_excel('jumlah sampah pertahun.xlsx', index=False)
```

```python
#SOAL NO 4
sampah_per_kota_tahun = {}

for index, row in df.iterrows():
    kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    sampah = row['jumlah_produksi_sampah']
    
    key = (kota, tahun) 
    if key in sampah_per_kota_tahun:
        sampah_per_kota_tahun[key] += sampah
    else:
        sampah_per_kota_tahun[key] = sampah

print("\nJawaban No.4 :")
print("Jumlah sampah per Kota/Kabupaten per Tahun:")
for key, total in sampah_per_kota_tahun.items():
    print(f"Kota: {key[0]}, Total Sampah: {total:.2f} ton")

#SIMPAN FILE
df_per_kota_tahun = pd.DataFrame(list(sampah_per_kota_tahun.items()), columns=['Kota/Tahun', 'Jumlah Sampah'])
df_per_kota_tahun.to_csv('hasil akhir.csv', index=False)
df_per_kota_tahun.to_excel('hasil akhir.xlsx', index=False)
```
